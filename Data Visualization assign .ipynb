```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn import datasets
%matplotlib inline
```

```python
# Loading the Iris Data
iris = datasets.load_iris()
```

```python
X_reduced = PCA(n_components = 3).fit_transform(iris.data)
X_reduced       # reduce dimension of data as we want 3 features to extract
y = iris.target # gives prediction or target  values
fig = plt.figure(1,figsize = (12,8)) 
ax = Axes3D(fig, elev=-150, azim=120)  # plot 3D plot with elevation and azimuthal angle given
ax.scatter(X_reduced[:,0],X_reduced[:,1],X_reduced[:,2], c = y, cmap = plt.cm.rainbow_r)
ax.set_title("iris tranformed 3d plot")
ax.set_xlabel("x data")
ax.set_ylabel("y data")
ax.set_zlabel("z data")
plt.show()
```
