### Problem Statement 
In this assignment students will build the random forest model after normalizing the variable to house pricing from boston data set.

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston
boston = load_boston()
X = pd.DataFrame(boston.data, columns=boston.feature_names) 
y = boston.target
```

```python
X.shape
```

```python
X.describe()
```

## Handling Missing Values

```python
X.isnull().sum()
```

No null values. Let's move ahead.


## Splitting The dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## Standardization

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Training The Model

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor()
regressor.fit(X_train, y_train)
```

## Predicting the Results

```python
y_pred = regressor.predict(X_test)
print("Score: ",regressor.score(X_test, y_test))
```

## Applying k-Fold Cross Validation

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = regressor, X = X_train, y = y_train, cv = 50)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

## Applying Grid Search to find the best model and the best parameters

```python
## This cell takes lots of time to execute!!
from sklearn.model_selection import GridSearchCV
# Create the parameter grid
param_grid = [{'n_estimators': np.arange(100,500,50),
               "criterion" : ["mse", "mae"],
               'max_depth': np.arange(10, 50, 10) ,
               'min_samples_split': np.arange(10, 50, 10),
               'min_samples_leaf': np.arange(10, 50, 10)}]




gs = GridSearchCV(estimator=regressor, param_grid=param_grid, verbose=2)

gs = gs.fit(X_train, y_train)

print(gs.best_score_)

print(gs.best_params_)

my_model = gs.best_estimator_

my_model.fit(X_train, y_train)
```

```python
y_pred = my_model.predict(X_test)

print("Accuraccuracy:", my_model.score(X_test, y_test))

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
print("R2 Score:",r2_score(y_test, y_pred))
print("mean_absolute_error:",mean_absolute_error(y_test, y_pred))
print("mean_squared_error:",mean_squared_error(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = my_model, X = X_train, y = y_train, cv = 50)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```
