## Problem Statement 

In this assignment students have to compress racoon grey scale image into 5 clusters. 

In the end, visualize both raw and compressed image and look for quality difference. 

The raw image is  available in spicy.misc package with the name face.

```python
import numpy as np
from sklearn.cluster import KMeans
from scipy import misc
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
face = misc.face(gray=True)
```

```python
# Let's see the Original Image.
plt.figure(1)
plt.imshow(face, cmap=plt.cm.gray, vmin=face.min(), vmax=face.max())
```

```python
n_clusters = 5
np.random.seed(0)

print("Original Image Shape: ",face.shape)
X = face.reshape((-1,1))
print("Resized Image Shape: ",X.shape)
```

```python
k_means = KMeans(n_clusters=5, n_init=4, verbose=3, random_state=0)
k_means.fit(X)
```

## Create The Values and Labels

```python
values = k_means.cluster_centers_.squeeze()
labels = k_means.labels_
```

## Create an Array for Labels and Values

```python
face_compressed = np.choose(labels, values)

face_compressed.shape = face.shape
```

```python
# Let's see the Original Image.
plt.figure(1)
plt.imshow(face, cmap=plt.cm.gray, vmin=face.min(), vmax=256)

## Compressed face
plt.figure(2)
plt.imshow(face, cmap=plt.cm.gray, vmin=face.min(), vmax=face.max())
```
